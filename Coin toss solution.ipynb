```python
import random
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize

win = [7, 3, 5, 6]
lose = [0, 1, 2, 4]

win_mul = 1.07
lose_mul = 0.90
```

```python
def trial():
    toss = random.randint(0, 7)
    if toss in win:
        return True
    else:
        return False

def play(funds, total_plays):
    Play_num = []
    Funds = []
    play = 0
    while play < total_plays:
        if trial():
            funds = funds * win_mul
            Play_num.append(play)
            Funds.append(funds)
        else:
            funds = funds * lose_mul
            Play_num.append(play)
            Funds.append(funds)
        play = play + 1
    plt.plot(Play_num,Funds)
    #print (Funds[-1])
    Final_funds.append(Funds[-1])
    return(Final_funds)
```

```python
Final_funds= []

end_single_random_variable = play(10, 100) # for a single random variable
figsize(25, 8)
plt.ylabel('Fund')
plt.xlabel('#bets')
plt.show()
print (Final_funds)
print (end_single_random_variable)
```
