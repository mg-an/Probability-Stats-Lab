```python
import random
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
#designed from the player's point of view for risk appetite
def Croupier():
    #let's take European Roulette's Impair et Pair
    ball = random.randint(0, 36) #37 for European, 38 for American
    if (random.randint(0, 1)) == ball % 2: #if you win the bet, which is also a random variable
        return True
    else:
        return False # for simpler case combining the house with the counter party
```

```python
def play(funds, wager, total_plays):
    Play_num = [] # for plotting
    Funds = []    # for plotting
    play = 1 #initialising
    while play < total_plays:
        #If we win, winning streak
        if Croupier():
            funds = funds + wager
            Play_num.append(play)
            Funds.append(funds)
        #If the counter party wins, losing streak
        else:
            funds = funds - wager
            Play_num.append(play)
            Funds.append(funds)
        play = play + 1 #updating counts
    plt.plot(Play_num,Funds)
    Final_funds.append(Funds[-1])
    return(Final_funds)
```
